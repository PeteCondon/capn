```python
import time

def sieve(n):
    sieve = [True] * n
    sieve[0:2] = [False, False]

    for i in range(2, int(n**0.5) + 1):
        if sieve[i]:
            for j in range(i*i, n, i):
                sieve[j] = False

    return [i for i, prime in enumerate(sieve) if prime]
```

```python
import math
import time
from itertools import chain

def composite_values(values, limit):
  result = set()
  stack = list(values)

  result_add = result.add
  result_contains = result.__contains__
  stack_append = stack.append

  while stack:
      x = stack.pop()
      for v in values:
          y = x * v
          if y >= limit:
              break

          if not result_contains(y):
              result_add(y)
              stack_append(y)
  return result

def primorial_sieve(x):
  p = [2,3,5]
  pn = [2,6]

  for i in range(2, x):
    pn.append(math.prod(p[0:i+1]))
    pni = pn[i]
    pni_1 = pn[i-1]
    t = p[i:]

    cv = composite_values(t, pni)

    t.insert(0, 1)
    t.extend([x for x in cv if x < pni_1])
    t.sort()
    cv_set = set(cv)

    px = [p]
    for j in range(1, p[i]):
      b = pni_1 * j
      px.append([x + b for x in t])

    p = [x for x in list(chain.from_iterable(px)) if x not in cv]
    p.sort()
  return p
```

```python
start = time.time()
sieve_primes = sieve(2*3*5*7*11*13*17*19)#*23)
end = time.time()

print(f"Found {len(sieve_primes)} primes in {end - start:.4f} seconds")
```
